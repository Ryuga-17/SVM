```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/cell_samples.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
for f in df.columns:
  print(df[f].value_counts())
```

```python
df=df[pd.to_numeric(df['BareNuc'],errors='coerce').notnull()]
df['BareNuc']=df['BareNuc'].astype('int')
```

```python
df.dtypes
```

```python
x=df.iloc[:,1:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn import svm
```

```python
cls = svm.SVC(kernel='linear', gamma='auto', C=2)
```

```python
cls.fit(x_train, y_train)
```

```python
cls.score(x_test,y_test)
```

```python
import pickle
with open('cls.pkl', 'wb') as f:
  pickle.dump(cls,f)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x)
p=scaler.transform(x)
```

```python
p=pd.DataFrame(p)
```

```python
x_train
```

```python
p.head()
```

```python
p_train,p_test,q_train,q_test=train_test_split(p,y,test_size=0.3)
```

```python
cls.fit(p_train, q_train)
cls.score(p_test,q_test)
```

```python
cls.predict(p_test)
```

```python
print((p.iloc[:,-1]).value_counts())
```

```python
from sklearn.pipeline import Pipeline

steps=([('scaler',MinMaxScaler()),('cls',svm.SVC(kernel='linear', gamma='auto', C=2))])
```

```python
pipe=Pipeline(steps)
pipe.fit(x_train,y_train)
```

```python
pipe.predict(x_test)
```

```python
pipe.score(x_test,y_test)
```

```python
from sklearn import set_config
set_config(display='diagram')
pipe
```

```python
import pickle
with open('Clspipe.pkl', 'wb') as f:
  pickle.dump(pipe,f)
```
